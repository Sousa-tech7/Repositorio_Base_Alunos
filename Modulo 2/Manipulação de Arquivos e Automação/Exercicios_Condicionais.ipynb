Classifique a nota do aluno:

- 5 até 6.9: "Regular"

- 0 até 4.9: "Reprovado"

mostre para o aluno a classificação dele através de um print, exemplo print("Excelente").

```python
nota = float(input("Qual a sua nota ? :"))
if nota >= 9.0 and nota <=10:
    print('Excelente')
elif nota  >= 7.0 and nota <=8.9:
    print('Bom')
elif nota >= 5.0 and nota <=6.9:   
    print('Regular')
else:
    print('Reprovado')

```

Exercício 8 - Dia da Semana
Peça um número de 1 a 7 e mostre o dia da semana correspondente: 📅

- 1: Domingo

- 2: Segunda

- ... até 7: Sábado

```python
numero = int(input("Digite um número de 1 até 7"))
if numero == 1:
    print('Domingo')
elif numero == 2:
    print('Segunda')
elif numero == 3:
    print('Terça')
elif numero == 4:
    print('Quarta')
elif numero == 5:
    print('Quinta')
elif numero == 6:
    print('Sexta')
elif numero == 7:
    print('Sábado')
else:
    print('Você colocou um dia ou um número que não existe')
```

Exercício 9 - Calculadora Simples 📏📐

- Faça uma calculadora que:

- Pede dois números

- Pede a operação (+, -, *, /)

- Dependendo da operação mostre o resultado.



```python
primeiro_numero = float(input('Digite um número:'))
segundo_numero = float(input('Digite um segundo número:'))
operacao = input('Digite a sua operacao (+,-,/,*)')

if operacao =="+":
    print(f'A soma dos números {primeiro_numero} e {segundo_numero} é igual {primeiro_numero + segundo_numero}')
elif operacao == "-":
    print(f'A subtração dos números {primeiro_numero} e {segundo_numero} é igual {primeiro_numero - segundo_numero}')
elif operacao == "*":
    print(f'A multiplicação dos números {primeiro_numero} e {segundo_numero} é igual {primeiro_numero * segundo_numero}')
elif operacao == "/":
    print(f'A divisão dos números {primeiro_numero} e {segundo_numero} é igual {primeiro_numero / segundo_numero} ') 
else:
    print("Error: sistema operacional inválido")               
```

Exercício 10 - Sistema de Multas 🚙🚘🧾


- Velocidade máxima: 80 km/h 🚫

- Até 80: "Sem multa"

- 81-100: "Multa leve"

- Acima de 100: "Multa grave"

```python
km = int(input('Digite a sua velocidade:'))
if km == 80:
    print('Sem multa')
elif km >= 81 and km <=100:
    print('Multa leve')
else:
    print('Multa grave')
```
